```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Import necessary NLP libraries
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("/content/all_kindle_review .csv")

# Display the first few rows of the DataFrame
df.head()
```

```python
#Removing punctuations, special characters and stopwords from the text column and  Converting the text to lower case.

import string

# Define function to remove punctuation and special characters
def remove_punctuation(text):
    return text.translate(str.maketrans("", "", string.punctuation))

# Define function to remove stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return " ".join(filtered_tokens)

# Apply preprocessing to the reviewText column
df["Cleaned_Text"] = df["reviewText"].apply(remove_punctuation)
df["Cleaned_Text"] = df["reviewText"].apply(remove_stopwords)
df["Cleaned_Text"] = df["Cleaned_Text"].str.lower()

# Display the modified DataFrame
df.head()
```

```python


#Creating two objects X and y. X will be the ' Cleaned_text' which is prepared in above row, and y will be the “Rating” column.
#Vectorizing ‘Cleaned_text’ using CountVectorizer
#and TF-IDF for bigrams and display the vectors

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Create X and y objects
X = df['Cleaned_Text']
y = df['rating']

# Vectorize Cleaned_Text using CountVectorizer for bigrams
count_vectorizer = CountVectorizer(ngram_range=(2, 2))
X_count = count_vectorizer.fit_transform(X)

# Vectorize Cleaned_Text using TF-IDF for bigrams
tfidf_vectorizer = TfidfVectorizer(ngram_range=(2, 2))
X_tfidf = tfidf_vectorizer.fit_transform(X)

# Display the vectors
print("CountVectorizer (Bigrams) Vectors:")
print(X_count.toarray())

print("\nTF-IDF (Bigrams) Vectors:")
print(X_tfidf.toarray())
```

```python
#Vectorizing ‘Cleaned_Text’ using Skip-Gram and CBOW using context words sizes [1,5] and vector size as 50.

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

from nltk.tokenize import word_tokenize

sentences = X.tolist()
# Tokenize the sentences
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]

# Define function to train Word2Vec model with different context window sizes
def train_word2vec_model(sentences, window_size, vector_size, sg):
    model = Word2Vec(sentences, sg=sg, window=window_size, vector_size=vector_size)
    return model

# Vectorize reviewText using Word2Vec with different context window sizes
context_sizes = [1, 2, 3, 4, 5]
vector_size = 50

for context_size in context_sizes:
    print(f"Word2Vec (Skip-Gram) Vectors with context window size = {context_size}:")
    model_sg = train_word2vec_model(sentences, context_size, vector_size, sg=1)

    print("Word\t\tVector")
    for word in model_sg.wv.index_to_key:
        vector = model_sg.wv[word]
        print(f"{word}\t\t{vector}")

    print(f"\nWord2Vec (CBOW) Vectors with context window size = {context_size}:")
    model_cbow = train_word2vec_model(sentences, context_size, vector_size, sg=0)

    print("Word\t\tVector")
    for word in model_cbow.wv.index_to_key:
        vector = model_cbow.wv[word]
        print(f"{word}\t\t{vector}")

    print("------------------------")
```

```python
#Displaying the embeddings and interpreting the results for predicting similar
#words to 'fantastic'as  target word using the results of embeddings on different window sizes
#used in above task

import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Create a list of sentences
sentences = df['Cleaned_Text'].apply(lambda x: word_tokenize(x.lower())).tolist()

# Define target word for similarity comparison
target_word = 'fantastic'  # Replace with any target word you want to explore

for context_size in context_sizes:
    print(f"Word2Vec Vectors with context window size = {context_size}:")
    model = train_word2vec_model(sentences, context_size, vector_size, sg=1)

    print(f"Similar words to '{target_word}':")
    similar_words = model.wv.most_similar(target_word)
    for word, similarity in similar_words:
        print(f"{word}: {similarity}")

    print("------------------------")
```
